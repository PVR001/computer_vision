**1. Conversion from one color model to another**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
srcimg = cv2.imread('img.png')
srcimgs = cv2.cvtColor(srcimg,cv2.COLOR_BGR2RGB)
plt.imshow(srcimgs)
```

```python
plt.figure(figsize=(50, 50), constrained_layout=False)
BGR2HSV= cv2.cvtColor(srcimgs,cv2.COLOR_BGR2HSV )
```

```python
BGR2YUV=cv2.cvtColor(srcimgs,cv2.COLOR_BGR2YUV )
BGR2HLS=cv2.cvtColor(srcimgs,cv2.COLOR_BGR2HLS )
BGR2YCrCb=cv2.cvtColor(srcimgs,cv2.COLOR_BGR2YCrCb )
```

```python
plt.subplot(121),plt.imshow(srcimgs)
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(BGR2HSV)
plt.title('Output image'), plt.xticks([]), plt.yticks([])
plt.show()
```

***2. Include different types of noises in the input image with various densities ***

1.Gaussian Blur (Gaussian smoothing)

```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt


srcimgs = cv2.cvtColor(srcimg,cv2.COLOR_BGR2RGB)
oimg = srcimgs
gaussBlur =cv2.GaussianBlur(oimg,(7,7),cv2.BORDER_DEFAULT)
plt.figure(figsize=(50, 50), constrained_layout=False)
plt.subplot(161), plt.imshow(oimg), plt.title("Input Image")
plt.subplot(162), plt.imshow(gaussBlur), plt.title("ImageafterGaussianSmoothing")
```

1.Mean filter

```python
import numpy as np
import pandas as pd
import cv2

import matplotlib.pyplot as plt

srcimgs = cv2.cvtColor(srcimg,cv2.COLOR_BGR2RGB)
oimg = srcimgs
kernel = np.ones((3,3),np.float32)/25
meanFilter = cv2.filter2D(oimg,-1,kernel)
#cv2.imshow("Mean filtered Image",np.hstack((img, meanFilter)))

plt.figure(figsize=(50, 50), constrained_layout=False)
plt.subplot(161), plt.imshow(oimg), plt.title("Input Image")
plt.subplot(162), plt.imshow(meanFilter), plt.title("Image after MeanFilter")
```

2.High pass filter

```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt

srcimgs = cv2.cvtColor(srcimg,cv2.COLOR_BGR2RGB)
oimg = srcimgs
gaussBlur =cv2.GaussianBlur(oimg,(9,9),cv2.BORDER_DEFAULT)
#cv2.imshow("Gaussian Smoothing",np.hstack((oimg,gaussBlur)))

plt.figure(figsize=(50,50), constrained_layout=False)
plt.subplot(161), plt.imshow(oimg), plt.title("Input Image")
plt.subplot(162), plt.imshow(gaussBlur), plt.title("Gaussian filtered Image")

hpf=cv2.subtract(oimg, gaussBlur)#oimg – gaussBlur
# or We can use this statement to filter the high pass image
plt.subplot(163), plt.imshow(hpf), plt.title("Gaussian filtered Image")
```

3.low pass filter

```python
srcimgs = cv2.cvtColor(srcimg,cv2.COLOR_BGR2RGB)
oimg = srcimgs
kernel = np.ones((3,3),np.float32)/9
mf = cv2.filter2D(oimg,-1, kernel)
lpf = cv2.subtract(oimg, mf)
plt.figure(figsize=(50,50), constrained_layout=False)
plt.subplot(161), plt.imshow(oimg), plt.title("Input Image")
plt.subplot(162), plt.imshow(mf), plt.title("ImageafterMF")
plt.subplot(163), plt.imshow(lpf), plt.title("Image after Lowpass filter")
```

```python
srcimgs = cv2.cvtColor(srcimg,cv2.COLOR_BGR2RGB)
oimg = srcimgs
kernel = np.ones((5,5),np.float32)/9
mf = cv2.filter2D(oimg,-1, kernel)
lpf = cv2.subtract(oimg, mf)
plt.figure(figsize=(50,50), constrained_layout=False)
plt.subplot(161), plt.imshow(oimg), plt.title("Input Image")
plt.subplot(162), plt.imshow(mf), plt.title("ImageafterMF")
plt.subplot(163), plt.imshow(lpf), plt.title("Image after Lowpass filter")
```

```python
srcimgs = cv2.cvtColor(srcimg,cv2.COLOR_BGR2RGB)
oimg = srcimgs
kernel = np.ones((7,7),np.float32)/9
mf = cv2.filter2D(oimg,-1, kernel)
lpf = cv2.subtract(oimg, mf)
plt.figure(figsize=(50,50), constrained_layout=False)
plt.subplot(161), plt.imshow(oimg), plt.title("Input Image")
plt.subplot(162), plt.imshow(mf), plt.title("ImageafterMF")
plt.subplot(163), plt.imshow(lpf), plt.title("Image after Lowpass filter")
```

4.Laplacian filter

```python
srcimgs = cv2.cvtColor(srcimg,cv2.COLOR_BGR2RGB)
oimg = srcimgs
#img = cv2.cvtColor(oimg,cv2.COLOR_BGR2RGB)

LaplacianFilter = cv2.Laplacian(oimg, 24, (3,3))

plt.figure(figsize=(50, 50), constrained_layout=False)
plt.subplot(161), plt.imshow(oimg), plt.title("Input Image")
plt.subplot(162), plt.imshow(LaplacianFilter), plt.title("ImageafterLaplacianFilter")
```

5.Bilateral filter
